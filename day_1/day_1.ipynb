```python
import regex as re
with open("input_1", "r") as f:
    file = f.read()
#     file = """two1nine
# eightwothree
# abcone2threexyz
# xtwone3four
# 4nineeightseven2
# zoneight234
# 7pqrstsixteen
# """
    input = file.splitlines()
    total = 0
    for line in input:
        # print(line)
        numbers = re.findall(r"[0-9]|one|two|three|four|five|six|seven|eight|nine", line, overlapped=True)
        # numbers_1 = re.findall(r"[0-9]|one|two|three|four|five|six|seven|eight|nine", line, overlapped=False)
        # if len(numbers) != len(numbers_1):
        #     print(line)
        #     print("here")
        #     print(numbers)
        #     print(numbers_1)

        # print(numbers)
        for i in range(len(numbers)):
            numbers[i] = numbers[i]
            if numbers[i] == "one":
                numbers[i] = "1"
            elif numbers[i] == "two":
                numbers[i] = "2"
            elif numbers[i] == "three":
                numbers[i] = "3"
            elif numbers[i] == "four":
                numbers[i] = "4"
            elif numbers[i] == "five":
                numbers[i] = "5"
            elif numbers[i] == "six":
                numbers[i] = "6"
            elif numbers[i] == "seven":
                numbers[i] = "7"
            elif numbers[i] == "eight":
                numbers[i] = "8"
            elif numbers[i] == "nine":
                numbers[i] = "9"
            # else:
            #     print(numbers[i])
        if len(numbers) ==0:
            print("No numbers found")
        if len(numbers) == 1:
            number = int(numbers[0]+numbers[0])
            # print("here")
        else:
            number = int(numbers[0]+numbers[-1])
        # print(number)
        total += number    
print(total)
```
